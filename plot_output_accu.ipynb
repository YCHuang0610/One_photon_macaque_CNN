```python
import numpy as np

layers_2 = np.load('output/correlation_rows_MyCNN_two_layers.npy')
layers_4 = np.load('output/correlation_rows_MyCNN_four_layers.npy')
layers_6 = np.load('output/correlation_rows_MyCNN_six_layers.npy')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 将layers_2、layers_4、layers_6这三个不同模型结果的准确度用箱线图绘制在同一张图上
plt.figure(figsize=(10, 6))
sns.boxplot(data=[layers_2, layers_4, layers_6], width=0.5)
plt.xticks([0, 1, 2], ['2 layers', '4 layers', '6 layers'])
plt.ylabel('Correlation coefficient')
plt.title('Correlation coefficient of different models')
plt.savefig('plots/accu_of_different_models.png')
```
